```python
from bs4 import BeautifulSoup
import requests 
```

```python
url = "https://boston.craigslist.org/d/housing/search/hhh"
```

```python
response = requests.get(url)
```

```python
response
```

```python
data = response.text
```

```python
data
```

```python
soup = BeautifulSoup(data,'html.parser')
```

```python
tags = soup.find_all('a')
```

```python
for tag in tags:
    
    print(tag.get('href'))
    
```

```python
titles = soup.find_all("a",{"class":"result-title"})
for title in titles:
    print(title.text)
```

```python
adresses = soup.find_all("span",{"class":"result-hood"})
```

```python
for adress in adresses: 
    print(adress.text)
    
```

```python
prices = soup.find_all("span",{"class":"result-price"})
```

```python
for price in prices:
    print(price.text)
```

```python
descriptions = soup.find_all('p',{'class':'result-info'})
for description in descriptions:
    title = description.find('a',{'class':'result-title'}).text
    adress_tag = description.find('span',{'class':'result-hood'})
    adresse = adress_tag.text[2:-1] if adress_tag else 'N/A'
    price = description.find('span',{'class':'result-price'}).text
    date = description.find('time',{'class':'result-date'}).text
    link = description.find('a',{'class':'result-title'}).get('href')
    print('Description title:', title, '\nAdress:',adress.text,'\nDate', date, '\nLink:',link,'price:',price,'\n--')
   
```
